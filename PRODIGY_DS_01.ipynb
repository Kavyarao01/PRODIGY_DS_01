# Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.

```python
#Import Libraries

import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
#Read data set

df = pd.read_csv("Employees.csv")
```

```python
df=pd.DataFrame(df)
y=df['AbsentHours']
x=df['Gender']
pr= plt.figure(figsize=(40, 15))
plt.xlabel("Gender",fontweight='bold',fontsize=20)
plt.ylabel("AbsentHours",fontweight='bold',fontsize=20)
plt.xticks(fontweight='bold',rotation='vertical',fontsize=20)
plt.bar(x,y, color='tomato',align='center')
     
```

```python
plt.figure(figsize=(12,10))
plt.title('AbsentHours')
sb.histplot(x='AbsentHours', hue="DepartmentName", data=df)
plt.show()
```

```python
plt.figure(figsize=(12,10))
plt.title('AbsentHours')
sb.histplot(x='AbsentHours', hue="City", data=df)
plt.show()
```
